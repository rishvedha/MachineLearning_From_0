```python
import torch
tokens=["I play","cricket","every","day"]
token_ids=[10350,2045,1023,100]
embeded_tokens=torch.randn(4,2048)
print(embeded_tokens.shape)
```

```python
w_q=torch.randn(2048,2048)
w_k=torch.randn(2048,2048)
w_v=torch.randn(2048,2048)
```

```python
q=embeded_tokens@w_q
k=embeded_tokens@w_k
v=embeded_tokens@w_v
```

```python
print(q.shape,k.shape,v.shape)
```
